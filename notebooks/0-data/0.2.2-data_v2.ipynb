## Libraries

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Read data

```python
df = pd.read_csv('../../data/processed/processed_data.csv')
```

```python
## Final processing
# Drop file columns
df.drop(columns='file',inplace=True)
# group by sexo and drop sex column
data_m = df[df.sexo==1].drop(columns="sexo")
data_f = df[df.sexo==2].drop(columns="sexo")

# divido cada perimetro por la altura
```

```python
# divido cada masa por el peso de la persona
for masa in ['masa_piel','masa_adiposa', 'masa_muscular', 'masa_residual', 'masa_osea']:
    data_m[masa] = data_m[masa] / data_m['peso']
    data_f[masa] = data_f[masa] / data_f['peso']
```

```python
# calculo el porcentaje total. Este número es distinto de 1, tiene un leve error.
data_m['peso_porcentaje'] = data_m[['masa_piel','masa_adiposa', 'masa_muscular', 'masa_residual', 'masa_osea']].sum(axis=1)
data_f['peso_porcentaje'] = data_f[['masa_piel','masa_adiposa', 'masa_muscular', 'masa_residual', 'masa_osea']].sum(axis=1)
data_m
```

```python
# vuelvo a dividir cada masa por el "peso_porcentaje" para hacer que la suma de 1
for masa in ['masa_piel','masa_adiposa', 'masa_muscular', 'masa_residual', 'masa_osea']:
    data_m[masa] = data_m[masa] / data_m['peso_porcentaje']
    data_f[masa] = data_f[masa] / data_f['peso_porcentaje']
```

```python
data_m[['masa_piel','masa_adiposa', 'masa_muscular', 'masa_residual', 'masa_osea']].sum(axis=1)
```

```python

# Normalizado
mean_m = data_m.mean()
data_m.to_csv('../../data/interim/mean_m_v2.csv', index=False)
std_m = data_m.std()
data_m.to_csv('../../data/interim/std_m_v2.csv', index=False)

mean_f = data_f.mean()
data_m.to_csv('../../data/interim/mean_f_v2.csv', index=False)
std_f = data_f.std()
data_m.to_csv('../../data/interim/std_f_v2.csv', index=False)

inputs = data_m.columns.drop(['masa_piel',
       'masa_adiposa', 'masa_muscular', 'masa_residual', 'masa_osea','peso_porcentaje'])

for columna in inputs:
    data_m[columna] = (data_m[columna] - data_m[columna].mean()) / data_m[columna].std()
    data_f[columna] = (data_f[columna] - data_f[columna].mean()) / data_f[columna].std()
```
