## Extrayendo data

```python
columnas = ["file","sexo","edad","peso","talla","per_brazo_rel","per_brazo_ten","per_antebrazo",
"per_torax","per_cintura","per_cadera","per_muslo_max","per_muslo_medial",
"per_pantorrilla","masa_piel","masa_adiposa","masa_muscular","masa_residual",
"masa_osea"]
df = pd.DataFrame(columns=columnas)
```

```python
sheet = "Proc datos brutos"
for file in files:
    try:
        temp_df = pd.read_excel(io=os.path.join('../../data/raw',file), sheet_name=sheet)
        #generales
        sexo = temp_df.iloc[0,8]
        edad = temp_df.iloc[1,8]
        peso = temp_df.iloc[4,1]
        talla = temp_df.iloc[5,1]
        #perimetros
        per_brazo_rel = temp_df.iloc[16,1]
        per_brazo_ten = temp_df.iloc[17,1]
        per_antebrazo = temp_df.iloc[18,1]
        per_torax = temp_df.iloc[19,1]
        per_cintura = temp_df.iloc[20,1]
        per_cadera = temp_df.iloc[21,1]
        per_muslo_max = temp_df.iloc[22,1]
        per_muslo_medial = temp_df.iloc[23,1]
        per_pantorrilla = temp_df.iloc[24,1]
        #masas resultado
        masa_piel = temp_df.iloc[45,1]
        masa_adiposa = temp_df.iloc[50,1]
        masa_muscular = temp_df.iloc[60,1]
        masa_residual = temp_df.iloc[66,1]
        masa_osea = temp_df.iloc[76,1]
        
        lista = [file,sexo,edad,peso,talla,per_brazo_rel,per_brazo_ten,per_antebrazo,per_torax,
                per_cintura,per_cadera,per_muslo_max,per_muslo_medial,per_pantorrilla,
                masa_piel,masa_adiposa,masa_muscular,masa_residual,masa_osea
                ]

        row = pd.Series(lista, index=df.columns)
        df = df.append(row,ignore_index=True)
    except:
        pass
    
```

## Visualizando data

```python
for column in df.columns:
    if column != 'file':
        plt.figure();
        #sns.histplot(data=df,x=column, hue='sexo', palette='inferno');  
        sns.kdeplot(data=df,x=column, shade=True, hue='sexo', palette=['b','r'])
```

```python
for column in df.columns:
    if column != 'file':
        plt.figure();
        sns.catplot(data=df, y=column, x='sexo',kind='box', palette=['b','r'])
```

```python

for column in df.columns:
    if column != 'file':
        plt.figure();
        sns.scatterplot(data=df, x=column, y='masa_muscular', hue='sexo', palette=['b','r'])
```
